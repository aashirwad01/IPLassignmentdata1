## LISTING OF WIN PERCENTAGE OF VARIOUS IPL TEAMS

```python
i=0
mth_awon=list()
while(i<14):
    print(mth['winner'].value_counts().sort_index(axis=0,inplace=False)[i:i+1])
    mth_awon.append(mth['winner'].value_counts().sort_index(axis=0,inplace=False)[i]/mth_t.sort_index(axis=0,inplace=False)[i])
    print(mth['winner'].value_counts().sort_index(axis=0,inplace=False)[i]/mth_t.sort_index(axis=0,inplace=False)[i])
    i=i+1
```

```python
for i in range(len(mth_awon)):
    print(mth_awon[i])
len(mth_awon)
```
